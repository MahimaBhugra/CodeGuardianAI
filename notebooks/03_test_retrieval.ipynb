# Testing Retrieval

In this notebook, we test the retrieval component of the RAG pipeline.

Workflow:

User Query
↓
Embedding Search
↓
ChromaDB
↓
Relevant Documents

The retrieved context will later be sent to DeepSeek-Coder for code review.

```python
import chromadb
```

```python
#This code creates a PersistentClient for ChromaDB, which will store our vector database in the "chroma_db" directory. If the directory doesn't exist, it will be created automatically. We then get the collection named "code_review_knowledge" from the database, which will be used to store our coding standards, security guidelines, and review examples as vectors. Finally, we print a success message to confirm that we've connected to the database successfully.
client = chromadb.PersistentClient(
    path="../chroma_db"
)

collection = client.get_collection(
    name="code_review_knowledge"
)

print("Connected Successfully")
```

```python
query = "division by zero"

results = collection.query(
    query_texts=[query],
    n_results=2
)

print(results["documents"][0])
```

```python
query = "SQL Injection"

results = collection.query(
    query_texts=[query],
    n_results=2
)

print(results["documents"][0])
```
