# Building the Vector Database

## Objective

This notebook creates the vector database required for Retrieval-Augmented Generation (RAG).

### Workflow

Data Files
↓
Sentence Transformer Embeddings
↓
ChromaDB Vector Store
↓
Semantic Retrieval

The stored knowledge will later be retrieved and supplied to DeepSeek-Coder during code reviews.

```python
import chromadb
from sentence_transformers import SentenceTransformer

print("Imports Successful")
```

```python
#This model converts text into Embeddings(numerical representations of text) that can be used for similarity search and other NLP tasks.
model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)

print("Embedding Model Loaded")
```

```python
#This code creates our Vector Database using ChromaDB. It will be stored in the "chroma_db" directory. If it doesn't exist, it will be created.
client = chromadb.PersistentClient(
    path="../chroma_db"
)

collection = client.get_or_create_collection(
    name="code_review_knowledge"
)

print("ChromaDB Connected")
```
